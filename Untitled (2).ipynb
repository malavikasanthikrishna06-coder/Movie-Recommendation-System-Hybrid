```python
import pandas as pd
import numpy as np
import warnings
# Suppress the numpy sparsity runtime warnings to keep our output clean
warnings.filterwarnings('ignore', category=RuntimeWarning)

print("📦 1. Downloading official GroupLens MovieLens-100k dataset...")
!wget -q -nc http://files.grouplens.org/datasets/movielens/ml-100k.zip
!unzip -q -n ml-100k.zip

print("📥 2. Parsing and structuring data matrices...")
movies_path = 'ml-100k/u.item'
ratings_path = 'ml-100k/u.data'

# Parse full movie features sheet
movies_cols = ['movie_id', 'title', 'release_date', 'video_release_date', 'IMDb_URL',
               'unknown', 'Action', 'Adventure', 'Animation', 'Childrens', 'Comedy',
               'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror',
               'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
movies_df = pd.read_csv(movies_path, sep='|', names=movies_cols, encoding='latin-1')

# Parse ratings
ratings_cols = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings_df = pd.read_csv(ratings_path, sep='\t', names=ratings_cols)

# Build baseline interaction frames
movie_data = pd.merge(ratings_df[['user_id', 'movie_id', 'rating']], movies_df[['movie_id', 'title']], on='movie_id')
movie_matrix = movie_data.pivot_table(index='user_id', columns='title', values='rating')

# Aggregate statistical metrics per film
ratings_stats = movie_data.groupby('title').agg(
    average_rating=('rating', 'mean'),
    number_of_ratings=('rating', 'count')
).reset_index()

print("⚙️ 3. Deploying Hybrid Recommendation Engine Architecture...")
def get_hybrid_recommendations(movie_title, min_ratings_threshold=60):
    try:
        # Isolate active genres for content matching
        target_movie_genres = movies_df[movies_df['title'] == movie_title].iloc[0, 5:]
        active_genres = target_movie_genres[target_movie_genres == 1].index.tolist()

        # Calculate matrix similarity (Collaborative Filtering)
        target_movie_ratings = movie_matrix[movie_title]
        similar_to_target = movie_matrix.corrwith(target_movie_ratings)

        corr_df = pd.DataFrame(similar_to_target, columns=['Correlation']).dropna()
        corr_summary = corr_df.join(ratings_stats.set_index('title'))
        candidates = corr_summary[corr_summary['number_of_ratings'] >= min_ratings_threshold].reset_index()

        # Enforce content boundaries (Content-Based Filtering)
        genres_expanded = movies_df[['title'] + active_genres]
        valid_genre_titles = genres_expanded[genres_expanded[active_genres].sum(axis=1) > 0]['title'].tolist()

        hybrid_rec = candidates[candidates['title'].isin(valid_genre_titles)]
        hybrid_rec = hybrid_rec.sort_values('Correlation', ascending=False)

        print(f"\n🎬 Base Film: '{movie_title}' | Detected Genre Categories: {active_genres}")
        return hybrid_rec.head(6)

    except Exception as e:
        return f"❌ Film title not found or error occurred: {e}"

print("\n🔥 4. Testing End-to-End Execution Flow...")
results = get_hybrid_recommendations('L.A. Confidential (1997)')
print("\n🏆 Top Smart Portfolio Recommendations:")
print(results[['title', 'Correlation', 'average_rating']].to_string(index=False))
```
